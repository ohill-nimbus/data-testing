```python
clients_df = portal_client.query_to_pandas(
    """WITH RECURSIVE category_path AS (
    SELECT id, name, parent_id, 0 AS depth
    FROM api_clients
    WHERE parent_id = 1
    AND deleted_at IS NULL

    UNION ALL

    SELECT c.id, c.name, c.parent_id, cp.depth + 1
    FROM api_clients c
    INNER JOIN category_path cp ON cp.id = c.parent_id
    AND deleted_at IS NULL
    )
    SELECT * FROM category_path;"""
).set_index("id")
```

```python
devices_df = portal_client.query_to_pandas(
    """SELECT DISTINCT user_id, device_type_id
    FROM topaz.api_devices
    WHERE device_type_id IN (4, 5)
    AND active
    ORDER BY user_id, device_type_id"""
)
```
